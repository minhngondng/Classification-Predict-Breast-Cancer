```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('breast_cancer.csv')
df
```

```python
 df['diagnosis'] = df['diagnosis'].apply(lambda text: 1 if text == "M" else 0)
```

```python
df = df.drop('Unnamed: 32', axis = 1)
df = df.drop('id', axis = 1)
```

```python
df.describe().transpose()
```

```python
sns.countplot(data = df, x = 'diagnosis')
```

```python
df.corr()['diagnosis'][:-1].sort_values().plot(kind = 'bar')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X = df.drop('diagnosis', axis = 1).values
y = df['diagnosis'].values
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 101)
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_train.shape
```

```python
model = Sequential()
```

```python
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))


#BINARY CLASSIFICATION
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
## Early Stopping
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))


#BINARY CLASSIFICATION
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose = 3, patience = 25)
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
new_loss = pd.DataFrame(model.history.history)
new_loss.plot()
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))



#BINARY CLASSIFICATION
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
new_losses_2 = pd.DataFrame(model.history.history)
```

```python
new_losses_2.plot()
```

```python
predictions = (model.predict(X_test) > 0.5).astype("int32")
```

```python
predictions
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```
